# Sparse inverse covariance estimation

```python
model = GraphLassoCV(max_iter=2000)
model.fit(X.values)
cov_ = model.covariance_
vmax = cov_.max()
plt.imshow(cov_, interpolation='nearest', vmin=-vmax, vmax=vmax,
               cmap=plt.cm.RdBu_r)
plt.xticks(())
plt.yticks(())
```

Clustering features

```python
_, labels = affinity_propagation(cov_)
print labels
```

```python
cov_new = zeros((len(labels), len(labels)))
j = 0
for i in argsort(labels):
    cov_new[j] = cov_[i]
    j += 1
cov_new = cov_new.T
cov_n = zeros((len(labels), len(labels)))
j = 0
for i in argsort(labels):
    cov_n[j] = cov_new[i]
    j += 1
vmax=cov_n.max()
plt.imshow(cov_n.T, interpolation='nearest', vmin=-vmax, vmax=vmax,
               cmap=plt.cm.RdBu_r)
plt.xticks(())
plt.yticks(())
```
