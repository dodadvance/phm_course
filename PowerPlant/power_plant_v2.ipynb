#### Comparing coefs

```python
def coef(data, labels):
    lr = LinearRegression()
    lr.fit(data, labels)
    return lr.coef_
```

#### Active power

```python
# Normal data
norm_coefs_a = coef(X1, Y2.values)
# Abnrmal data
abnorm_coefs_a = coef(X1_test, Y2_test.values)
# Seminormal data
seminorm_coefs_a = coef(X1_semi, Y2_semi.values)
```

#### Reactive power

```python
# Normal data
norm_coefs_r = coef(X1, Y1.values)
# Abnrmal data
abnorm_coefs_r = coef(X1_test, Y1_test.values)
# Seminormal data
seminorm_coefs_r = coef(X1_semi, Y1_semi.values)
```

#### Plot coefs

```python
def plot_coefs(coefs1, coefs2, label1, label2, ttl, n_indexes):
#     sort_indexes = argsort(-abs(coefs1-coefs2) / abs(coefs1*coefs2))[:n_indexes]
    sort_indexes = argsort(-abs(coefs1-coefs2))[:n_indexes]
    bar_width = 0.5
    index = np.arange(n_indexes)
    linear_bar = bar(index, coefs1[sort_indexes], bar_width,
                     alpha=0.4,
                     color='b',
                     label=label1)
    ridge_bar = bar(index + bar_width, coefs2[sort_indexes], bar_width,
                     alpha=0.4,
                     color='r',
                     label=label2)
    title(ttl)
    xticks(index + bar_width, sort_indexes, rotation=45)
    legend(loc='best')
```

```python
subplot(1,3,1)
plot_coefs(norm_coefs_a, norm_coefs_r, 'Active', 'Reactive', 'Normal data', 7)
subplot(1,3,2)
plot_coefs(abnorm_coefs_a, abnorm_coefs_r, 'Active', 'Reactive', 'Abnormal data', 7)
subplot(1,3,3)
plot_coefs(seminorm_coefs_a, seminorm_coefs_r, 'Active', 'Reactive', 'Seminormal data', 7)
plot_filename = 'coef_plots_ar.png'
plt.savefig(plot_filename, format='png')
print 'Plot is saved to %s' % os.path.join(os.getcwd(), plot_filename)
```

```python
def plot_dif_coefs(coefs1, coefs2, coefs3, label1, label2, ttl, n_indexes):
#     sort_indexes = argsort(-abs(coefs1-coefs2) / abs(coefs1*coefs2))[:n_indexes]
#     sort_indexes = argsort(-abs(coefs1-coefs2))[:n_indexes]
    sort_indexes = argsort(-abs(coefs3-coefs2))[:n_indexes]
    bar_width = 0.5
    index = np.arange(len(sort_indexes))
    linear_bar = bar(index, coefs1[sort_indexes]-coefs2[sort_indexes], bar_width,
                     alpha=0.4,
                     color='b', label=label1)
    ridge_bar = bar(index + bar_width, coefs1[sort_indexes]-coefs3[sort_indexes], bar_width,
                     alpha=0.4,
                     color='r',
                     label=label2)
    title(ttl)
    xticks(index + bar_width, sort_indexes, rotation=45)
    legend(loc='best')
    return sort_indexes
```

```python
# Difference in coefs between normal and abnormal or seminormal data
figsize(20,10)
subplot(1,2,1)
ind1 = plot_dif_coefs(norm_coefs_a, abnorm_coefs_a, seminorm_coefs_a, 'Abnormal', 'Seminormal', 'Active power',5)
# Difference in coefs between normal and seminormal data
subplot(1,2,2)
ind2 = plot_dif_coefs(norm_coefs_r, abnorm_coefs_r, seminorm_coefs_r, 'Abnormal', 'Seminormal', 'Reactive power',5)
plot_filename = 'coef_diff.png'
savefig(plot_filename, format='png')
print 'Plot is saved to %s' % os.path.join(os.getcwd(), plot_filename)
```
